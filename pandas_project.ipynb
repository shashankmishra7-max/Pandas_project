```python
df1=pd.read_csv("/content/Student.csv")
df1
```

```python
print("Average mark of student \n",df1.mean())
```

```python
df1['Total_Score'] = df1['Maths'] + df1['Physics'] + df1['Chemistry']
top_10 = df1.sort_values(by='Total_Score', ascending=False).head(10)
print(top_10)
print("\n" + "="*40 + "\n")
```

```python
highest_scores = df1[['Maths', 'Physics', 'Chemistry']].max()
print(highest_scores)
print("\n" + "="*40 + "\n")
```

```python
def assign_grade(score):
    if score >= 80:
        return 'A'
    elif score >= 60:
        return 'B'
    elif score >= 40:
        return 'C'
    else:
        return 'D'

df1['Average_Score'] = df1['Total_Score'] / 3
df1['Grade'] = df1['Average_Score'].apply(assign_grade)
print(df1[['Maths', 'Physics', 'Chemistry', 'Total_Score', 'Average_Score', 'Grade']].head())
print("\nGRADE DISTRIBUTION COUNTS:")
print(df1['Grade'].value_counts())
print("\n" + "="*40 + "\n")
```

```python
summary_stats = df1[['Maths', 'Physics', 'Chemistry', 'Total_Score', 'Average_Score']].describe()
print(summary_stats)
```

```python
df2=pd.read_csv("/content/employee_salary_dataset.csv")
df2
```

```python

dept_avg_salary = df2.groupby('Department')['Monthly_Salary'].mean().reset_index()
print(dept_avg_salary)



highest_paid = df2.loc[df2['Monthly_Salary'].idxmax()]
lowest_paid = df2.loc[df2['Monthly_Salary'].idxmin()]

print("Highest Paid:")
print(highest_paid[['EmployeeID', 'Name', 'Department', 'Monthly_Salary']])
print("\nLowest Paid:")
print(lowest_paid[['EmployeeID', 'Name', 'Department', 'Monthly_Salary']])

gender_salary_stats = df2.groupby('Gender')['Monthly_Salary'].agg(['mean', 'median', 'min', 'max']).reset_index()
print(gender_salary_stats)

print("Average Salary by Education Level:")
edu_trend = df2.groupby('Education_Level')['Monthly_Salary'].mean().sort_values(ascending=False).reset_index()
print(edu_trend)

correlation_exp = df2['Experience_Years'].corr(df2['Monthly_Salary'])
correlation_age = df2['Age'].corr(df2['Monthly_Salary'])
print(f"\nCorrelation between Experience and Salary: {correlation_exp:.2f}")
print(f"Correlation between Age and Salary: {correlation_age:.2f}")

salary_report = df2.groupby(['City', 'Department'])['Monthly_Salary'].describe()
print(salary_report)
```
